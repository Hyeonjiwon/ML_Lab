```python
import numpy as np
import pandas as pd

datapath = 'data/ionosphere.data'

#['f0','f1',...,'f32'] + ['class']
raw_data = pd.read_csv(datapath, sep=',',names=(['f{}'.format(x) 
                                                 for x in range(0, 33)] + ['class']))

raw_data.head()
```

```python
from sklearn import preprocessing

le_class = preprocessing.LabelEncoder()
final_data = raw_data.copy()
final_data['class'] = le_class.fit_transform(raw_data['class'])

final_data.head()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split

features = ['f{}'.format(x) for x in range(0, 33)]

train_d, test_d = train_test_split(final_data, test_size=0.1)

train_y = train_d['class']
train_x = train_d[features]

test_y = test_d['class']
test_x = test_d[features]

model = MLPClassifier(max_iter=500)
model.fit(train_x, train_y)

# 결과가 다르게 나온이유 : train_test_split()상에서 suffle되기 때문
# 파라메터 초기값에 따라 달라짐
model.score(test_x, test_y)
```
